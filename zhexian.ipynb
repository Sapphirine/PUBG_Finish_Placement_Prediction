```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
train = pd.read_csv('train_V2.csv')
```

```python
data = train.copy()
f,ax1 = plt.subplots(figsize =(20,10))
sns.set_style("darkgrid")
sns.pointplot(x='kills',y='winPlacePerc',data=data,color='#8B0000',alpha=0.8)
plt.xlabel('Kills',fontsize = 15)
plt.ylabel('Percentage of Win',fontsize = 15)
plt.show()
```
