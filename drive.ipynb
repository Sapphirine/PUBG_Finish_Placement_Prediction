```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
train = pd.read_csv('train_V2.csv')
```

```python
data = train.copy()
print("{} players ({:.4f}%) drived for 0 meters. This means that they don't have a driving licence yet.".format(len(data[data['rideDistance'] == 0]), 100*len(data[data['rideDistance']==0])/len(train)))
```

```python
data = train.copy()
plt.figure(figsize=(15,10))
plt.title("Fig of Riding Distance",fontsize=15)
sns.distplot(data['rideDistance'],color='#9ACD32')
plt.show()
plt.figure(figsize=(15,10))
plt.title("Fig of Riding Distance",fontsize=15)
sns.distplot(data['rideDistance'],hist=False,color='#9ACD32')
plt.show()
plt.figure(figsize=(15,10))
plt.title("Fig of Riding Distance",fontsize=15)
sns.distplot(data['rideDistance'],kde=False,color='#9ACD32')
plt.show()
```

```python
sns.jointplot(x="winPlacePerc", y="rideDistance", data=train, ratio=3, color="#9ACD32",kind="reg")
plt.show()
```

```python
data = train.copy()

data['rideDistance'] = pd.cut(data['rideDistance'], [-1, 0, 500, 1000, 3000,5000,48390], labels=['No Ride','1~100m', '100~1000m','1000~3000m','3000~5000m', 'Higher than 5000m'])

plt.figure(figsize=(15,8))
sns.boxplot(x="rideDistance", y="winPlacePerc", data=data,palette="pastel")
plt.show()
```
