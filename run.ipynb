```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
train = pd.read_csv('train_V2.csv')
```

```python
print("The average person walks for {:.1f}m, 99% of people have walked {}m or less, while the marathoner champion walked for {}m.".format(train['walkDistance'].mean(), train['walkDistance'].quantile(0.99), train['walkDistance'].max()))
```

```python
data = train.copy()
plt.figure(figsize=(15,10))
plt.title("Fig of Walking Distance",fontsize=15)
sns.distplot(data['walkDistance'],color='plum')
plt.show()
plt.figure(figsize=(15,10))
plt.title("Fig of Walking Distance",fontsize=15)
sns.distplot(data['walkDistance'],hist=False,color='plum')
plt.show()
plt.figure(figsize=(15,10))
plt.title("Fig of Walking Distance",fontsize=15)
sns.distplot(data['walkDistance'],kde=False,color='plum')
plt.show()
```

```python
sns.jointplot(y="winPlacePerc", x="walkDistance", data=train, ratio=3, color="plum",kind='reg')
plt.show()
```

```python
data = train.copy()

data['walkDistance'] = pd.cut(data['walkDistance'], [-1, 0, 500, 1000, 3000,5000,17300], labels=['No Run','1~100m', '100~1000m','1000~3000m','3000~5000m', 'Higher than 5000m'])

plt.figure(figsize=(15,8))
sns.boxplot(x="walkDistance", y="winPlacePerc", data=data,palette="pastel")
plt.show()
```
