```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
train = pd.read_csv('train_V2.csv')
```

```python
train.plot(x="kills",y="winPlacePerc", kind="scatter", figsize = (15,10))
```

```python
sns.jointplot(x='winPlacePerc', y='matchDuration', data=train, ratio=3, color='r')
plt.show()
```

```python
data = train.copy()
plt.figure(figsize=(15,10))
plt.title('Match duration Distribution',fontsize=15)
sns.distplot(data['matchDuration'])
plt.show()
```
