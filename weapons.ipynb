```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
train = pd.read_csv('train_V2.csv')
```

```python
data = train.copy()
# data.loc[data['kills'] > data['kills'].quantile(0.995)] = 'less than 0.5%'
plt.figure(figsize=(15,5))
sns.countplot(data['weaponsAcquired'].sort_values(),palette="Set3")
plt.title("Fig of Weapons Acquired",fontsize=20)
plt.show()
```

```python
data = train.copy()
data1 = data.loc[data['weaponsAcquired'] < data['weaponsAcquired'].quantile(0.995)]
plt.figure(figsize=(15,5))
sns.countplot(data1['weaponsAcquired'].sort_values(),palette="coolwarm")
plt.title("Fig of Weapons Acquired",fontsize=20)
plt.show()
```

```python
data = train.copy()
data1 = data.loc[data['weaponsAcquired'] < data['weaponsAcquired'].quantile(0.9995)]
plt.figure(figsize=(15,5))
sns.countplot(data1['weaponsAcquired'].sort_values(),palette="coolwarm")
plt.title("Fig of Weapons Acquired",fontsize=20)
plt.show()
```

```python
data = train.copy()
data = data.loc[data['weaponsAcquired'] < data['weaponsAcquired'].quantile(0.9995)]
plt.figure(figsize=(15,10))
plt.title("weaponsAcquired",fontsize=15)
sns.distplot(data['weaponsAcquired'],color="tan") 
plt.show()
plt.figure(figsize=(15,10))
plt.title("weaponsAcquired",fontsize=15)
sns.distplot(data['weaponsAcquired'],hist=False,color="tan") 
plt.show()
plt.figure(figsize=(15,10))
plt.title("weaponsAcquired",fontsize=15)
sns.distplot(data['weaponsAcquired'],kde=False,color="tan") 
plt.show()
```

```python
sns.jointplot(x="winPlacePerc", y="weaponsAcquired", data=train, ratio=3, color="tan",kind='reg')
plt.show()
```

```python
data = train.copy()

data['weaponsAcquired'] = pd.cut(data['weaponsAcquired'], [-1, 0, 3, 6, 10,15,236], labels=['No Weapon','1~3 Weapons', '4~6 Weapons','7~10 Weapons','10~15 Weapons','More than 15 weapons'])

plt.figure(figsize=(15,8))
sns.boxplot(x="weaponsAcquired", y="winPlacePerc", data=data,palette="pastel")
plt.show()
```
